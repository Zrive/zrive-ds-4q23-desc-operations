```python
import pandas as pd
from transformers import BartTokenizer, BartForConditionalGeneration
```

```python

# load and token the model
model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)

data= pd.read_csv( "data/POC Description of operations - Sheet3_FINAL.csv"
)

# Add new column
data["summary"] = None

# Go through each row
for index, row in data.head(2).iterrows():
    description = str(row['WEB'])

    # Token ang generate resume
    inputs = tokenizer(description, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(**inputs, max_length=150, min_length=50, length_penalty=2.0, num_beams=4)
    summary_text = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    # Update
    data.at[index, 'summary'] = summary_text

# Final csv
data.to_csv("data/POC Description of operations - RESULT.csv", index=False)
```
