```python
import pandas as pd
from transformers import pipeline

# Cargar el modelo de resumen
summarizer = pipeline("summarization", model="Azma-AI/bart-large-text-summarizer")

# Leer el conjunto de datos desde el archivo CSV
data = pd.read_csv("/home/pamalo9/zrive-ds-4q23-desc-operations/src/data/POC Description of operations - Sheet3_FINAL.csv")

# Añadir una nueva columna para los resúmenes
data["summary_azma_ai"] = None

# Iterar sobre las dos primeras filas y generar los resúmenes
for index, row in data.head(2).iterrows():
    description = str(row['WEB'])
    summary = summarizer(description, max_length=150, min_length=50, length_penalty=2.0, num_beams=4)
    
    # Imprimir el diccionario devuelto por el modelo
    print(summary)

    # Acceder al resumen en el diccionario (ajusta según la estructura)
    data.at[index, 'summary_azma_ai'] = summary[0].get('summary_text', summary[0].get('text', ''))

# Guardar el DataFrame con los resúmenes en un nuevo archivo CSV
data.to_csv("/home/pamalo9/zrive-ds-4q23-desc-operations/src/data/POC Description of operations - RESULT2.csv", index=False)
```
